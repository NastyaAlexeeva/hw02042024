```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных о стоимости недвижимости
house_data = pd.read_csv("kc-house-data.csv", encoding="utf-8")

# Подсчет количества домов с видом на набережную и без
waterfront_counts = house_data["waterfront"].value_counts()

# Построение столбчатой диаграммы
plt.figure(figsize=(8, 6))
waterfront_counts.plot(kind="bar", color=["skyblue", "orange"])
plt.title("Распределение домов от наличия вида на набережную")
plt.xlabel("Наличие вида на набережную")
plt.ylabel("Количество домов")
plt.xticks([0, 1], ["Без вида на набережную", "С видом на набережную"], rotation=0)
plt.grid(axis="y")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных о стоимости недвижимости
house_data = pd.read_csv("kc-house-data.csv", encoding="utf-8")

# Построение ящика с усами для возраста дома и цены недвижимости
plt.figure(figsize=(10, 6))
house_data.boxplot(column="price", by="yr_built")
plt.title("Влияние возраста дома на цену недвижимости")
plt.xlabel("Возраст дома")
plt.ylabel("Цена недвижимости")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Гистограмма площади жилых помещений
plt.figure(figsize=(10, 6))
plt.hist(house_data["sqft_living"], bins=20, color="skyblue", edgecolor="black")
plt.title("Распределение площади жилых помещений")
plt.xlabel("Площадь жилых помещений (кв. футы)")
plt.ylabel("Частота")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Ящик с усами для количества спален и цены недвижимости
plt.figure(figsize=(10, 6))
plt.boxplot(house_data["price"], vert=False)
plt.title("Влияние количества спален на цену недвижимости")
plt.xlabel("Цена")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
# Ящик с усами для вид на воду и цены недвижимости
plt.figure(figsize=(10, 6))
plt.boxplot([house_data[house_data["waterfront"] == 0]["price"], house_data[house_data["waterfront"] == 1]["price"]], labels=["Нет", "Да"])
plt.title("Влияние наличия вид на воду на цену недвижимости")
plt.xlabel("Вид на воду")
plt.ylabel("Цена")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
# Столбчатая диаграмма для этажности и цены недвижимости
plt.figure(figsize=(10, 6))
plt.bar(house_data["floors"], house_data["price"], color="skyblue")
plt.title("Влияние этажности на цену недвижимости")
plt.xlabel("Этажность")
plt.ylabel("Цена недвижимости")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
condition_counts = house_data["condition"].value_counts()
plt.pie(condition_counts, labels=condition_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Распределение состояния домов")
plt.axis('equal')  
plt.show()
```
